### Plotting the data 

```python
df.spx.plot(title='S&P500 Prices',figsize=(20,5),label="S&P")
df.ftse.plot(title='FTSE100 Prices',figsize=(20,5),label="FTSE")
plt.legend()
plt.title("S&P VS FTSE")
```

### QQ plot

```python
import scipy.stats
import pylab
```

```python
#first param is the series we want  to test
#the second one is the way we wan to express the plot
scipy.stats.probplot(df.spx,plot=pylab)
pylab.show()
```
