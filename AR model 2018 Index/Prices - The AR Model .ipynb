## Importing the relevant packages


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.arima.model import ARIMA
from scipy.stats.distributions import chi2
import statsmodels.tsa.stattools as sts 
import seaborn as sns
sns.set()
pd.options.mode.chained_assignment = None
```

## Importing the Data and Pre-processing 

```python
raw_csv_data = pd.read_csv("Index2018.csv") 
df_comp=raw_csv_data.copy()
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
df_comp.set_index("date", inplace=True)
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

```python
df_comp['market_value']=df_comp.ftse
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## The ACF

```python
sgt.plot_acf(df.market_value,zero=False,lags=40)#zero=False to exclude the AC with the first lag
plt.title("ACF Plot",size=20)
```

## The PACF

```python
sgt.plot_pacf(df.market_value,zero=False,lags=40,alpha=0.05,method='ols')#zero=False to exclude the AC with the first lag
plt.title("PACF Plot",size=20)
```

## The AR(1) Model

```python
model_ar=ARIMA(df.market_value,order=(1,0,0))
```

```python
results=model_ar.fit()
```

```python
results.summary()
```

## Higher-Lag AR Models

```python
model_ar2=ARIMA(df.market_value,order=(2,0,0))
```

```python
results2=model_ar2.fit()
```

```python
results2.summary()
```

```python
model_ar3=ARIMA(df.market_value,order=(3,0,0))
results3=model_ar3.fit()
results3.summary()
```

```python
model_ar4=ARIMA(df.market_value,order=(4,0,0))
results4=model_ar4.fit()
results4.summary()
```

## LLR Test

```python
def LLR_test(mod1,mod2,df=1):
    l1=mod1.fit().llf
    l2=mod2.fit().llf
    lr=2*(l2-l1)
    p=chi2.sf(lr,df).round(3)
    return p
    
```

## Comparing Higher-Lag AR Models

```python
LLR_test(model_ar2,model_ar3)
```

small p-value even for 5% significance level, we reject the null hypothesis that says the simple model is better.which means we hould test more complex models until the p-value can no longer be rejected.

```python
model_ar6=ARIMA(df.market_value,order=(6,0,0))
results6=model_ar6.fit()
results6.summary()
```

```python
model_ar7=ARIMA(df.market_value,order=(7,0,0))
results7=model_ar7.fit()
results7.summary()
```

```python
model_ar8=ARIMA(df.market_value,order=(8,0,0))
results8=model_ar8.fit()
results8.summary()
```

```python
LLR_test(model_ar7,model_ar8)
```

The AR(7) is the best fir for the prices

## Analysing the Residuals

```python
df['res_prices']=results7.resid
```

```python
df.res_prices.mean()
```

```python
df.res_prices.var()
```

```python
sts.adfuller(df.res_prices)
```

we got stationnarity of resiuduals with mean around 0 but hih variance which indicates that the model didn't capture non stationnary behaviors, we can chech for ACF coefficients to see if residuals are basically white noise :

```python
sgt.plot_acf(df.res_prices,zero=False,lags=40)#zero=False to exclude the AC with the first lag
plt.title("ACF Plot of Residuals",size=20)
```

No autocorrelation is captured, which means the residuals are basically white noise
